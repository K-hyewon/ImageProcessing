```python
import cv2
import requests
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros((256,256,3), np.uint8)

img = cv2.line(img, (0,0), (120,50), (0,0,255), 5)
img = cv2.rectangle(img, (0,0), (120,120), (0,255,0), 5)
img = cv2.circle(img, (120,120), 20, (255,0,0), 3)
img = cv2.ellipse(img, (128,128), (80,40), -30, 0, 360, (255,255,255), 1)

points = np.array([[50,50], [150,50], [100,100]], np.int32)
img = cv2.polylines(img, [points], True, (255,0,255), 1)

font = cv2.FONT_HERSHEY_SIMPLEX
img = cv2.putText(img, 'Hyewon',(10,200), font, 1, (255,255,255), 2, cv2.LINE_AA)

plt.imshow(img[...,::-1], interpolation='bicubic')
plt.show()
```
