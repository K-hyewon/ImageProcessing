```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
from skimage import img_as_ubyte, img_as_float
```

# Translation

```python
cat = data.chelsea()
img = cv2.cvtColor(cat, cv2.COLOR_RGB2GRAY)

rows, cols = img.shape

M = np.float32([[1,0,150],
                [0,1,50]])
print(M)

dst = cv2.warpAffine(img, M, (cols, rows))

plt.figure(figsize=(12,12))
plt.subplot(121); plt.imshow(img, cmap='gray'), plt.axis('off')
plt.subplot(122); plt.imshow(dst, cmap='gray'), plt.axis('off')
plt.show()
```

# Rotation
- center  

```python
M = cv2.getRotationMatrix2D((cols/2, rows/2), 30, 1)
print(M)

dst = cv2.warpAffine(img, M, (cols, rows))
dst2 = cv2.warpAffine(img, M,)

plt.figure(figsize=(12,12))
plt.subplot(131); plt.imshow(img, cmap='gray'), plt.axis('off')
plt.subplot(132); plt.imshow(dst, cmap='gray'), plt.axis('off')
plt.subplot(133); plt.imshow(img, cmap='gray'), plt.axis('off')
plt.show()
```
